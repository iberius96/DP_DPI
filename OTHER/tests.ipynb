```python
def find_duplicate_rows(df, cols):
    # create a subdataframe with only the specified columns
    df_sub = df[cols]
    
    # create a boolean mask for the duplicate rows
    mask = df_sub.duplicated(keep=False)
    
    # get the indices of the duplicate rows
    duplicate_indices = df[mask].index
    
    return duplicate_indices

print(find_duplicate_rows(connections, ["BS", "FS name","FS macro region"]))
```

```python
def remove_character(df, column, character):
    # select the column as a series
    col = df[column]
    
    # replace the character with an empty string
    col_clean = col.str.replace(character, '')
    
    # assign the cleaned series back to the original dataframe
    df[column] = col_clean


remove_character(connections_url, 'BS', '(')
remove_character(connections_url, 'BS', ')')

df2 = connections_url.copy()

newdict = {}
for key, value in res_dict.items():
    for string in value:
        newdict.setdefault(string, []).append(key)

connections_url['BS'] = connections_url['BS'].map(newdict)
connections_url = connections_url.fillna(df2)

connections_url[['BS']] = connections_url[['BS']].astype(str)

remove_character(connections_url, 'BS', '[')
remove_character(connections_url, 'BS', "'")
remove_character(connections_url, 'BS', ']')

connections_url.to_csv("connection_temp.csv")
connections_url
```
